## ElGamal
$p$ primo, $r$ raiz primitiva de $p$, $1\le a \le p-1$, $b\equiv r^a \mod p$. <br>
ChPub: $(p, r, b)$ <br>
ChPriv: $a$

Cifrar $mens$: $(\gamma, \delta)$
$$\gamma = r^k \mod p$$
$$\delta = mens\cdot b^k \mod p$$

Decifrar: $$\left(\gamma^a\right)^{-1}\cdot \delta \mod p$$

```python
p = random_prime(2^256, lbound = 2^250)
Zp = IntegerModRing(p)
r = Zp.multiplicative_generator()
r
```

```python
a = randint(2, p-2)
a
```

```python
b = r^a
b
```

```python
ChPub = p, r, b
```

```python
mens = 1234
k = randint(2, p-2)
```

```python
gamma, delta = r^k, mens* b^k
gamma, delta
```

```python
1/(gamma^a)*delta
```
