### Primos de Mersènne

Se $d\mid n$ então $(2^d-1)\mid (2^n-1)$.

Logo, se $2^n-1$ é primo então $n$ é primo.

<br>

A $M_n = 2^n-1$ chama-se número de Mersènne. $M_n$ é um primo de Mersènne se $M_n$ é primo. Portanto, $M_n$ é um primo de Mersènne implica que $n$ seja primo.

### Teste de Lucas-Lehmer
Seja $p$ primo. $M_p$ é primo de Mersènne se e só se passa o teste de Lucas-Lehmer.

   $r_1 \equiv 4 \mod M_p$ <br>
   $r_k = r_{k-1}^2-2 \mod M_p$<br>
   $M_p$ passa o teste de Lucas-Lehmer se $r_{p-1} \equiv 0 \mod M_p$


```python
def Lucas_Lehmer(p):
    if not is_prime(p):
        return False
    Mp = 2^p-1
    r = mod(4, Mp)
    for k in range(1, p-1):
        r = r^2-2
    return r == 0
```
