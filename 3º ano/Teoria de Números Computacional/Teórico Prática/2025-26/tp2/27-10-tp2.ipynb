```python
def RSA_gen(nbits=1024):
    p = random_prime(2^(nbits//2), 2^(nbits//2-1))
    q = random_prime(2^(nbits//2+1), 2^(nbits//2))
    n = p*q
    m = (p-1)*(q-1)
    e = randint(2, m-1)
    while gcd(e, m) != 1:
        e = randint(2, m-1)
    d = power_mod(e, -1, m)
    pub_key = (n, e)
    priv_key = d
    return pub_key, priv_key

def RSA_encrypt(mens, PubK):
    n, e = PubK
    c = power_mod(mens, e, n)
    return c

def RSA_decrypt(cripto, chave):
    pub, priv = chave
    n, _ = pub
    decifr = power_mod(cripto, priv, n)
    return decifr
```
