## ElGamal

$p$ primo, $r$ raiz primitiva de $p$, $1\le a\le p-1$, $b  \equiv r^a \mod p$.

ChPub: $(p, r, b)$.

ChPriv: $a$.

Cifrar $mens$: $$(\gamma, \delta) = (r^k, mens\cdot b^k)$$

Decifrar: $$(\gamma^a)^{-1} \cdot \delta$$

```python
p = 113
Zp = IntegerModRing(p)
```

```python
r = Zp.multiplicative_generator()
a = randint(2, p-2)
b = r^a
```

```python
ChPub = p, r, b
ChPub
```

```python
mens = 78
k = randint(2, p-2)
gama, delta = r^k, mens*b^k
gama, delta
```

```python
1/gama^a*delta
```
