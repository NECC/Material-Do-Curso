### 5. Factorial
Ask the user for a number `n` and compute `n!` (factorial) using a loop (not recursion).

- $1! = 1$
- $2! = 2 \times 1 = 2$
- $3! = 3 \times 2 \times 1 = 6$
- $4! = 4 \times 3 \times 2 \times 1 = 24$
- $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$

```python
#using for
n = int(input("n (>= 0): ").strip())

if n < 0:
    print("Factorial is not defined for negative numbers.")
else:
    res = 1
    for i in range(2, n + 1):
        res *= i
    print(f"{n}! = {res}")


    ##  or
    #res = 1
    #for i in range(n, 0, -1):
    #  res *= i


    ##  or
    #res = 1
    #for i in range(1, n + 1):
    #  res *= i
```

```python
#using while
n = int(input("n (>= 0): ").strip())

if n < 0:
    print("Factorial is not defined for negative numbers.")
else:
    res = 1
    i = 2
    while i <= n:
        res *= i
        i += 1
    print(f"{n}! = {res}")
```

### 1. Palindrome
Write `is_palindrome(word)` returning `True` if `word` reads the same backwards.  

```python
def is_palindrome(word):
    s = str(word).strip().lower()
    return s == s[::-1]

print(is_palindrome("Radar"), is_palindrome("Python"))
```

### 2. Normalize
Write a function `normalize(xs)` that takes a list of numbers and returns a new list scaled to sum to 1.
- If the sum is 0, return a list of zeros of the same length.

```python
def normalize(xs):
    s = sum(xs)
    if s == 0:
        return [0 for _ in xs]
    return [x / s for x in xs]

print(normalize([1, 1, 2]))   # [0.25, 0.25, 0.5]
print(normalize([0, 0, 0]))   # [0, 0, 0]
```

### 3. Frequency table
Implement `freq(s)` returning a dict `{char: count}` for a string.

```python
def freq(s: str) -> dict:
    out = {}
    for ch in s:
        out[ch] = out.get(ch, 0) + 1
    return out

print(freq("abracadabra"))
```

### 4. — List/dict comprehension
Given `records = [('Ana', 15), ('Bruno', 12), ('Carla', 18)]`, build:
1) a dict `grades`
2) a list of names with grade >= 14

```python
records = [('Ana', 15), ('Bruno', 12), ('Carla', 18)]

# 1) dict name -> grade
grades = {name: g for name, g in records}

# 2) names with grade >= 14
approved = [name for name, g in records if g >= 14]

print(grades)
print(approved)
```

### 1. Input/Output

A) Create a file called `temperatures.txt` with temperature data (one temperature per line), then read it back and calculate the average.

B) Write a function that appends a new measurement to an existing file.

```python
from statistics import mean

# 1) Create a file with temperatures (one per line), read back, compute average
temps = [21.5, 22.0, 20.8, 23.1, 22.7]
fname = "temperatures.txt"

# Write
with open(fname, "w", encoding="utf-8") as f:
    for t in temps:
        f.write(f"{t}\n")

# Read + average (robust to stray whitespace)
values = []
with open(fname, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if line:
            values.append(float(line))
print("Average temperature:", mean(values))

# 2) Append a new measurement to an existing file
def append_measurement(filename: str, value: float) -> None:
    with open(filename, "a", encoding="utf-8") as f:
        f.write(f"{value}\n")

append_measurement(fname, 24.0)
```

### 2. JSON

A) Create a dictionary representing a lab sample with properties like `name`, `pH`, `temperature`, and `date_collected`.

B) Save this dictionary as a JSON file called `sample.json`.

C). Read the JSON file back and print the sample information in a formatted way.

```python
import json
from datetime import date

# 1) Dictionary representing a lab sample
sample = {
    "name": "Sample A",
    "pH": 7.2,
    "temperature": 25.0,
    "date_collected": str(date.today()),
}

# 2) Save to JSON
with open("sample.json", "w", encoding="utf-8") as f:
    json.dump(sample, f, ensure_ascii=False, indent=2)

# 3) Read back and print nicely
with open("sample.json", "r", encoding="utf-8") as f:
    loaded = json.load(f)

print(f"Sample '{loaded['name']}' — pH={loaded['pH']}, T={loaded['temperature']}°C, collected={loaded['date_collected']}")
```

### 3. Try/Except
A) Write a function that converts a string to float, handling the case where the string is not a valid number.

B) Create a safe division function that handles division by zero.

```python
# 1) Safe string→float
def to_float(s: str, default=None):
    try:
        return float(s)
    except (TypeError, ValueError):
        return default

print(to_float("3.14"), to_float("abc", default="invalid"))

# 2) Safe division
def safe_div(a, b, default=None):
    try:
        return a / b
    except ZeroDivisionError:
        return default

print(safe_div(10, 2), safe_div(10, 0, default="∞"))
```

### 4. Recursion
A) Write a recursive function to calculate the power of a number: `power(base, exponent)`.

$$
2^4 = 2 \times 2 \times 2 \times 2 = 16
$$


B) Write a recursive function that counts down from a given number to 1, printing each number.

```python
def power(base: float, exponent: int) -> float:
    if exponent == 0:
        return 1
    if exponent == 1:
        return base
    return base * power(base, exponent - 1)
print(power(2, 4))
```

### 5. Classes: Library Book Management

**Problem:**

Create a Python class named `Book` to represent books in a library. Each `Book` object should have the following attributes:

- `title` (string)
- `author` (string)
- `isbn` (string, a unique identifier for the book)
- `is_available` (boolean, `True` if the book is available, `False` otherwise, default to `True`)

The `Book` class should also have the following methods:

- `__init__(self, title, author, isbn)`: Initializes the book with the given title, author, and ISBN. `is_available` should be set to `True` by default.
- `borrow_book(self)`: Changes the `is_available` status to `False` if the book is currently available. It should print a message indicating whether the book was successfully borrowed or if it's already on loan.
- `return_book(self)`: Changes the `is_available` status to `True` if the book is currently on loan. It should print a message indicating whether the book was successfully returned or if it was already available.
- `__str__(self)`: Returns a string representation of the book in the format: `"<Title> by <Author> (ISBN: <ISBN>) - Status: <Available/On Loan>"`

**Task:**

1. Define the `Book` class with the attributes and methods described above.
2. Create at least two `Book` objects.
3. Demonstrate the usage of `borrow_book`, `return_book`, and `__str__` methods by borrowing and returning books and printing their status.

```python
class Book:
    def __init__(self, title: str, author: str, isbn: str):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_available = True

    def borrow_book(self):
        if self.is_available:
            self.is_available = False
            print(f"'{self.title}' has been successfully borrowed.")
        else:
            print(f"'{self.title}' is currently on loan and cannot be borrowed.")

    def return_book(self):
        if not self.is_available:
            self.is_available = True
            print(f"'{self.title}' has been successfully returned.")
        else:
            print(f"'{self.title}' was already available.")

    def __str__(self):
        status = "Available" if self.is_available else "On Loan"
        return f"{self.title} by {self.author} (ISBN: {self.isbn}) - Status: {status}"


# Create Book objects
book1 = Book("The Lord of the Rings", "J.R.R. Tolkien", "978-0618260274")
book2 = Book("Pride and Prejudice", "Jane Austen", "978-0141439518")
book3 = Book("1984", "George Orwell", "978-0451524935")

# Demonstrate usage
print("Initial Status:")
print(book1)
print(book2)
print(book3)
print("\n---")

print("Attempting to borrow book1 and book2:")
book1.borrow_book()
book2.borrow_book()
print("\n---")

print("Current Status:")
print(book1)
print(book2)
print(book3)
print("\n---")

print("Attempting to borrow book1 again (should fail):")
book1.borrow_book()
print("\n---")

print("Attempting to return book1 and book3:")
book1.return_book()
book3.return_book() # book3 was never borrowed, should indicate already available
print("\n---")

print("Final Status:")
print(book1)
print(book2)
print(book3)
print("\n---")
```
