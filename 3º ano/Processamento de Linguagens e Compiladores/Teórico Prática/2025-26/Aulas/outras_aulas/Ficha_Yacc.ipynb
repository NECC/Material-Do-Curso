### 1. Lista de compras

Constroi um analisador sintático capaz de ler uma lista de compras, organizada por secções de um supermercado (CARNE, PEIXE, LEGUMES, BEBIDAS, etc.), e de validar sintaticamente essa lista.

Exemplo de uma lista de compras em que cada produto necessário de cada secção tem um **número de código, nome, preço unitário** e **quantidade a comprar** é apresentado a seguir:

---
```
CARNE :
  - 1 :: Bife :: 10.00 :: 1;
  - 2 :: Panado :: 5.00 :: 4;
  - 3 :: Hambúrguer :: 8.00 :: 3;
  - 4 :: Almôndegas :: 7.00 :: 5;

BEBIDA :
  - 5 :: Água :: 0.40 :: 16;
  - 6 :: Sumo :: 1.50 :: 9;
  - 7 :: Refrigerante :: 1.80 :: 10;

FRUTA :
  - 8 :: Maçã :: 0.60 :: 20;
  - 9 :: Banana :: 0.50 :: 15;
  - 10 :: Laranja :: 0.80 :: 18;
  - 11 :: Pêssego :: 0.70 :: 22;
  - 12 :: Uva :: 0.90 :: 17;

LEGUMES :
  - 13 :: Alface :: 1.00 :: 25;
  - 14 :: Tomate :: 0.75 :: 23;
  - 15 :: Cebola :: 0.50 :: 28;
  - 16 :: Batata :: 0.30 :: 30;
  - 17 :: Cenoura :: 0.40 :: 26;

PASTELARIA :
  - 18 :: Bolo de Chocolate :: 3.50 :: 1;
  - 19 :: Croissant :: 1.20 :: 14;
  - 20 :: Pastel de Nata :: 1.00 :: 5;
  - 21 :: Donut :: 0.80 :: 13;
```
---
Numa segunda etape, o processador a construir a partir da GIC escrita na fase anterior, deve calcular, para além do **custo total das compras**, o **número de items** listados, e o **total de artigos** a comprar.

Numa última etape, pretende-se que enriqueça o processador anterior com um Analisador Semântico  que detete três tipos de erros de inconsistência lógica da lista de compras e que só apresente os valores calculados se a frase for *bem reconhecida* (isto é, *se fizer sentido*). Os erros semânticos a detetar são:
+ nomes de secções repetidos;
+ mais do que uma referência ao mesmo produto  (mesmo código, nome e preço);
+ uso do mesmo código para produtos diferentes (com nomes ou preços unitários diversos).

Abaixo mostra-se  uma frase com todos estes tipos de erros semânticos:

```
CARNE :
    - 1 :: Bife :: 10.00 :: 4;
    - 1 :: Bife :: 10.00 :: 1;
    - 2 :: Lombo :: 10.00 :: 4;
    - 2 :: Lombo :: 12.00 :: 1;
    - 3 :: Figado :: 2.00 :: 1;

CARNE :
    - 3 :: Costoleta :: 2.00 :: 1;
    - 1 :: Bife :: 10.00 :: 7;

PEIXE :
    - 77 :: Pescada :: 20.00 :: 2;
```

### 4. Árvores binárias

Considera que uma árvore binária pode ser representada segundo um formato textual em que `()` representa uma árvore vazia e `( e l r )` representa uma árvore com raiz `e`, uma árvore `l` à sua esquerda e uma árvore `r` à sua direita. 

Por exemplo, `( 5 (3 (1 () ()) ()) (8 () (10 () ())))` representa uma árvore de raiz 5, com uma árvore de raiz 3 à sua esquerda e uma árvore de raiz 8 à sua direita.

### 6. Retro Art Typewriter

Há alguns anos atrás, as pessoas divertiam-se a criar "obras de arte" com os carateres das máquinas de escrever.
Mais tarde, essa tendência continuou com os primeiros computadores e os monitores tipo VT100.

Neste exercício, deverás processar um texto de entrada, contendo uma receita, e produzir uma destas imagens constituída apenas por carateres.

- Na receita, os blocos estão separados por espaço;
- Cada bloco irá indicar:
    1. `nl` - mudança de linha;
    2. `nx` - em que `n` será o número de vezes que deverás repetir o caráter `x`.

Exemplos:
- `4z` --> `zzzz`
- `1{` --> `{`
- `10=` --> `==========`
- `5bS` --> `\\\\\` (ver lista de abreviaturas mais abaixo)
- `27` --> `77`
- `123` --> `333333333333` (Se um bloco for composto apenas por digitos, o caráter será o último dígito)

Assim, se uma receita contiver:

```txt
2* 15sp 1x 4sQ nl
```

O resultado será:

```txt
**               x''''

```

Lista de abreviaturas a ser usada:
- `sp` = espaço
- `bS` = backSlash `\`
- `sQ` = singleQuote `'`
- `nl` = Mudança de linha 

Links para o tema:
- [Arte em carateres ASCII](https://asciiart.cc)
- [Livro sobre o tema com imagens interessantes](https://loriemerson.net/2013/01/18/d-i-y-typewriter-art/)
- [Vídeo sobre o tema](https://www.youtube.com/watch?v=kyK5WvpFxqo)

### Formato de input

String com a receita.

### Formato de Output

String contendo várias linhas mostrando a imagem que a receita cria.

### Restrições

- 5 <= Tamanho da receita <= 1000
- A receita não terá aspas no seu conteúdo;
- A receita deverá conter pelo meno 1 `nl`.

### Exemplo1 - input: um gatinho

```txt
1sp 1/ 1bS 1_ 1/ 1bS nl 1( 1sp 1o 1. 1o 1sp 1) nl 1sp 1> 1sp 1^ 1sp 1< nl 2sp 3|
```

### Exemplo1 - output: um gatinho 

```txt
 /\_/\
( o.o )
 > ^ <
  |||
```
